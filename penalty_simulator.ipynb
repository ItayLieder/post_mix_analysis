```python
# Cell 1: Imports
import soundfile as sf
import numpy as np
import pyloudnorm as pyln
from pydub import AudioSegment
import ffmpeg

# Cell 2: Load audio
filename = "mix_prep.wav"
data, rate = sf.read(filename)
meter = pyln.Meter(rate)  # BS.1770 meter
```

```python
# Cell 3: Measure loudness
loudness = meter.integrated_loudness(data)

# Approximate true peak (sample peak only, not oversampled true peak)
sample_peak = 20 * np.log10(np.max(np.abs(data)))
print(f"Integrated loudness: {loudness:.2f} LUFS")
print(f"Sample peak: {sample_peak:.2f} dBFS")
```

```python
# Cell 4: Compute penalty vs -14 LUFS target
target_lufs = -14.0
penalty_db = loudness - target_lufs
print(f"Expected platform penalty: -{penalty_db:.2f} dB")
```

```python
# Cell 5: Apply gain reduction to simulate normalization
gain_db = -penalty_db
normalized = data * (10 ** (gain_db / 20.0))
sf.write("simulated_normalized.wav", normalized, rate)
print("Saved normalized simulation -> simulated_normalized.wav")

# Cell 6: Simulate codec artifacts (AAC at 192k)
(
    ffmpeg
    .input("simulated_normalized.wav")
    .output("simulated_youtube.aac", acodec="aac", audio_bitrate="192k")
    .overwrite_output()
    .run()
)

# Decode back to WAV for listening comparison
(
    ffmpeg
    .input("simulated_youtube.aac")
    .output("simulated_youtube.wav")
    .overwrite_output()
    .run()
)
print("Saved YouTube simulation with AAC artifacts -> simulated_youtube.wav")
```

```python
import numpy as np
from scipy.io import wavfile
from scipy import signal

# --- Utilities ---
def db_to_lin(db):
    return 10.0 ** (db / 20.0)

def ensure_stereo(x):
    if x.ndim == 1:
        return np.stack([x, x], axis=-1)
    if x.shape[1] == 1:
        return np.repeat(x, 2, axis=1)
    return x

def butter_highpass(x, sr, cutoff_hz=20.0, order=2):
    nyq = 0.5 * sr
    norm = cutoff_hz / nyq
    b, a = signal.butter(order, norm, btype='highpass')
    return signal.lfilter(b, a, x, axis=0)

def peak_normalize(x, target_db=-6.0, eps=1e-12):
    peak = np.max(np.abs(x)) + eps
    gain = db_to_lin(target_db) / peak
    return x * gain

def to_float32(x, dtype):
    if dtype == np.int16:
        return x.astype(np.float32) / 32768.0
    if dtype == np.int32:
        return x.astype(np.float32) / 2147483648.0
    if dtype == np.uint8:
        return (x.astype(np.float32) - 128.0) / 128.0
    return x.astype(np.float32)

# --- Main function for notebooks ---
def prep_mix_for_landr(input_wav, output_wav, hpf_hz=20.0, target_peak_db=-6.0):
    sr, data = wavfile.read(input_wav)
    x = to_float32(data, data.dtype)
    x = ensure_stereo(x)

    # Step 1: remove DC & sub rumble
    x = butter_highpass(x, sr, cutoff_hz=hpf_hz, order=2)

    # Step 2: peak normalize for headroom
    x = peak_normalize(x, target_db=target_peak_db)

    # Safety clip
    x = np.clip(x, -1.0, 1.0)

    # Save as float32 WAV
    wavfile.write(output_wav, sr, x.astype(np.float32))
    print(f"âœ… Prepped mix saved to {output_wav} (SR={sr}, peak={np.max(np.abs(x)):.2f})")
```

```python
# A/B Master Comparison (Notebook-ready)
# - Compares two mastered WAVs: A = master of original mix, B = master of prepped mix
# - Outputs objective metrics and plots for an informed A/B
#
# Requirements: numpy, scipy, matplotlib (all standard in most Python stacks)
#
# Usage example (at the bottom of this cell):
# compare_masters("master_no_prep.wav", "master_with_prep.wav", title_a="No Prep", title_b="With Prep")
#
# What you get:
# - Metrics table (peak, RMS, DR proxy, bass/air balance, stereo width, correlation, spectral flatness)
# - Plots: spectrum up to 20 kHz, short-term RMS over time, waveform preview, cumulative loudness distribution
#
# Notes:
# - This avoids external loudness libs to stay dependency-light (not a full EBU R128 implementation).
# - Metrics are designed to be consistent and useful for relative comparison across the two masters.


import numpy as np
from scipy.io import wavfile
from scipy import signal
import matplotlib.pyplot as plt
import pandas as pd

try:
    from caas_jupyter_tools import display_dataframe_to_user
except Exception:
    display_dataframe_to_user = None


def _to_float32(x):
    if x.dtype == np.int16:
        return x.astype(np.float32) / 32768.0
    if x.dtype == np.int32:
        return x.astype(np.float32) / 2147483648.0
    if x.dtype == np.uint8:
        return (x.astype(np.float32) - 128.0) / 128.0
    return x.astype(np.float32)


def _ensure_stereo(x):
    if x.ndim == 1:
        return np.stack([x, x], axis=-1)
    if x.shape[1] == 1:
        return np.repeat(x, 2, axis=1)
    return x


def _resample_if_needed(x, sr, target_sr):
    if sr == target_sr:
        return x, sr
    gcd = np.gcd(sr, target_sr)
    up = target_sr // gcd
    down = sr // gcd
    y = signal.resample_poly(x, up, down, axis=0)
    return y, target_sr


def _db(x, eps=1e-12):
    return 20.0 * np.log10(np.maximum(eps, np.abs(x)))


def _short_term_rms_db(mono, sr, win_s=3.0, hop_s=0.5):
    win = int(win_s * sr)
    hop = int(hop_s * sr)
    if win <= 1:
        win = max(2, int(0.4 * sr))
        hop = max(1, int(0.1 * sr))
    # Use convolution for RMS
    kernel = np.ones(win) / win
    pow_sig = mono**2
    rms = np.sqrt(np.maximum(1e-20, np.convolve(pow_sig, kernel, mode="same")))
    # Downsample to hops
    idx = np.arange(0, len(mono), hop)
    rms_ds = rms[idx]
    rms_db = _db(rms_ds)
    t = idx / sr
    return t, rms_db


def _spectrum(mono, sr, n_fft=1<<16):
    seg = mono[:min(len(mono), n_fft)]
    if len(seg) < 2048:
        # zero-pad a short signal for a stable FFT
        pad = np.zeros(2048, dtype=seg.dtype)
        pad[:len(seg)] = seg
        seg = pad
    win = np.hanning(len(seg))
    sp = np.fft.rfft(seg * win)
    freqs = np.fft.rfftfreq(len(seg), 1/sr)
    mag_db = _db(np.abs(sp))
    return freqs, mag_db


def _bass_air_ratios(freqs, mag_db):
    # Convert dB spectrum to linear power to compute ratios robustly
    power = (10.0**(mag_db/20.0))**2
    total = np.sum(power) + 1e-20
    bass = np.sum(power[freqs < 120.0])
    air = np.sum(power[freqs > 8000.0])
    return float(100.0 * bass / total), float(100.0 * air / total)


def _spectral_flatness(mono, sr):
    # Simple spectral flatness measure (geometric mean / arithmetic mean)
    n_fft = 1<<14
    seg = mono[:min(len(mono), n_fft)]
    win = np.hanning(len(seg))
    sp = np.abs(np.fft.rfft(seg * win)) + 1e-12
    geo = np.exp(np.mean(np.log(sp)))
    ari = np.mean(sp)
    return float(np.clip(geo / ari, 0.0, 1.0))


def _stereo_metrics(stereo):
    L = stereo[:, 0]
    R = stereo[:, 1]
    # Phase correlation (mean of normalized product)
    denom = np.maximum(1e-12, np.sqrt(L**2) * np.sqrt(R**2))
    corr = np.mean((L * R) / denom)
    # Width proxy: average |S| / (|M| + 1e-12)
    M = 0.5 * (L + R)
    S = 0.5 * (L - R)
    width = float(np.mean(np.abs(S)) / (np.mean(np.abs(M)) + 1e-12))
    return float(np.clip(corr, -1.0, 1.0)), width


def _metrics_for_file(path, target_sr=48000):
    sr, data = wavfile.read(path)
    x = _to_float32(data)
    x = _ensure_stereo(x)
    x, sr = _resample_if_needed(x, sr, target_sr)

    mono = np.mean(x, axis=1)

    peak = float(np.max(np.abs(mono)))
    peak_db = float(_db(peak))
    rms = float(np.sqrt(np.mean(mono**2)))
    rms_db = float(_db(rms))
    crest = float(peak_db - rms_db)

    t, rms_db_st = _short_term_rms_db(mono, sr, win_s=3.0, hop_s=0.5)
    # DR proxy: 95th - 10th percentile of short-term loudness
    p95 = float(np.percentile(rms_db_st, 95))
    p10 = float(np.percentile(rms_db_st, 10))
    dr_proxy = float(p95 - p10)

    freqs, mag_db = _spectrum(mono, sr, n_fft=1<<16)
    bass_pct, air_pct = _bass_air_ratios(freqs, mag_db)
    flatness = _spectral_flatness(mono, sr)

    corr, width = _stereo_metrics(x)

    return {
        "sr": sr,
        "peak_dbfs": peak_db,
        "rms_dbfs": rms_db,
        "crest_db": crest,
        "loudness_p95_db": p95,
        "loudness_p10_db": p10,
        "dr_proxy_db": dr_proxy,
        "bass_energy_%": bass_pct,
        "air_energy_%": air_pct,
        "spectral_flatness": flatness,
        "phase_correlation": corr,
        "stereo_width": width,
        "time_axis": t,
        "short_term_rms_db": rms_db_st,
        "freqs": freqs,
        "mag_db": mag_db,
        "wave_mono": mono,
    }


def compare_masters(path_a, path_b, title_a="A (No Prep)", title_b="B (With Prep)"):
    A = _metrics_for_file(path_a)
    B = _metrics_for_file(path_b)

    # Metrics table
    rows = []
    keys = [
        "peak_dbfs", "rms_dbfs", "crest_db", "dr_proxy_db",
        "bass_energy_%", "air_energy_%",
        "spectral_flatness", "phase_correlation", "stereo_width",
    ]
    for k in keys:
        rows.append({
            "Metric": k,
            title_a: A[k],
            title_b: B[k],
            "Î” (B - A)": B[k] - A[k]
        })
    df = pd.DataFrame(rows)
    if display_dataframe_to_user:
        display_dataframe_to_user("Master A/B Metrics", df)
    else:
        print(df.to_string(index=False))

    # Plots
    # 1) Spectrum overlay (up to 20 kHz)
    maskA = A["freqs"] <= 20000.0
    maskB = B["freqs"] <= 20000.0
    plt.figure()
    plt.plot(A["freqs"][maskA], A["mag_db"][maskA], label=title_a)
    plt.plot(B["freqs"][maskB], B["mag_db"][maskB], label=title_b)
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude (dB)")
    plt.title("Magnitude Spectrum (up to 20 kHz)")
    plt.legend()
    plt.show()

    # 2) Short-term RMS over time
    # Align length
    tA, tB = A["time_axis"], B["time_axis"]
    rA, rB = A["short_term_rms_db"], B["short_term_rms_db"]
    L = min(len(tA), len(tB))
    plt.figure()
    plt.plot(tA[:L], rA[:L], label=title_a)
    plt.plot(tB[:L], rB[:L], label=title_b)
    plt.xlabel("Time (s)")
    plt.ylabel("Short-term RMS (dBFS)")
    plt.title("Short-term Loudness (3s window)")
    plt.legend()
    plt.show()

    # 3) Waveform preview (first 10 seconds, mono)
    sr = A["sr"]  # both resampled to same SR
    dur = min(10 * sr, min(len(A["wave_mono"]), len(B["wave_mono"])))
    t = np.arange(int(dur)) / sr
    # Normalize for visualization only
    a_vis = A["wave_mono"][:int(dur)]
    b_vis = B["wave_mono"][:int(dur)]
    a_vis = a_vis / (np.max(np.abs(a_vis)) + 1e-9)
    b_vis = b_vis / (np.max(np.abs(b_vis)) + 1e-9)
    plt.figure()
    plt.plot(t, a_vis, label=title_a)
    plt.plot(t, b_vis, label=title_b)
    plt.xlabel("Time (s)")
    plt.ylabel("Normalized amplitude")
    plt.title("Waveform (first 10 s, visualization-normalized)")
    plt.legend()
    plt.show()

    # 4) Cumulative distribution of short-term loudness
    def cdf(x):
        xs = np.sort(x)
        ys = np.linspace(0, 1, len(xs))
        return xs, ys
    xA, yA = cdf(A["short_term_rms_db"])
    xB, yB = cdf(B["short_term_rms_db"])
    plt.figure()
    plt.plot(xA, yA, label=title_a)
    plt.plot(xB, yB, label=title_b)
    plt.xlabel("Short-term RMS (dBFS)")
    plt.ylabel("Cumulative fraction")
    plt.title("Loudness Distribution (CDF)")
    plt.legend()
    plt.show()

    return df

```

```python
# Notebook-ready bass enhancement utilities
# - low_shelf_biquad(): RBJ low-shelf filter
# - boost_bass(): applies HPF (cleanup) + optional low-shelf boost + optional dynamic bass (band compression) + peak headroom
# - analyze/plot helpers for quick A/B checks
#
# Usage example (after running this cell):
# y = boost_bass("mix.wav", "mix_bassier.wav", shelf_db=3.0, shelf_hz=80, dyn_bass=True)
# compare_two("mix.wav", "mix_bassier.wav")
#
# This aims for MUSICAL, not boomy. Start with 2â€“3 dB @ 70â€“90 Hz.


import numpy as np
from scipy.io import wavfile
from scipy import signal
import matplotlib.pyplot as plt

def _to_float32(x):
    if x.dtype == np.int16:
        return x.astype(np.float32) / 32768.0
    if x.dtype == np.int32:
        return x.astype(np.float32) / 2147483648.0
    if x.dtype == np.uint8:
        return (x.astype(np.float32) - 128.0) / 128.0
    return x.astype(np.float32)

def _ensure_stereo(x):
    if x.ndim == 1:
        return np.stack([x, x], axis=-1)
    if x.shape[1] == 1:
        return np.repeat(x, 2, axis=1)
    return x

def _db(x, eps=1e-12):
    return 20*np.log10(np.maximum(eps, np.abs(x)))

def highpass(x, sr, hz=20.0, order=2):
    b, a = signal.butter(order, hz/(sr*0.5), btype='highpass')
    return signal.lfilter(b, a, x, axis=0)

def lowshelf_biquad(x, sr, f0=80.0, gain_db=3.0, S=0.5):
    """
    RBJ low-shelf filter. S ~ slope (0.1..1 gentle, >1 steeper).
    """
    A = 10**(gain_db/40.0)
    w0 = 2*np.pi*f0/sr
    cosw0 = np.cos(w0); sinw0 = np.sin(w0)
    alpha = sinw0/2 * np.sqrt((A + 1/A)*(1/S - 1) + 2)
    b0 =    A*((A+1) - (A-1)*cosw0 + 2*np.sqrt(A)*alpha)
    b1 =  2*A*((A-1) - (A+1)*cosw0)
    b2 =    A*((A+1) - (A-1)*cosw0 - 2*np.sqrt(A)*alpha)
    a0 =        (A+1) + (A-1)*cosw0 + 2*np.sqrt(A)*alpha
    a1 =   -2*((A-1) + (A+1)*cosw0)
    a2 =        (A+1) + (A-1)*cosw0 - 2*np.sqrt(A)*alpha
    b = np.array([b0, b1, b2]) / a0
    a = np.array([1.0, a1/a0, a2/a0])
    return signal.lfilter(b, a, x, axis=0)

def band_low(x, sr, cutoff=120.0, order=4):
    b, a = signal.butter(order, cutoff/(sr*0.5), btype='lowpass')
    return signal.lfilter(b, a, x, axis=0)

def envelope_detector(mono, sr, attack_ms=10.0, release_ms=120.0):
    a_a = np.exp(-1.0/((attack_ms/1000.0)*sr))
    a_r = np.exp(-1.0/((release_ms/1000.0)*sr))
    env = np.zeros_like(mono)
    prev = 0.0
    for i, v in enumerate(np.abs(mono)):
        if v > prev:
            prev = a_a*prev + (1-a_a)*v
        else:
            prev = a_r*prev + (1-a_r)*v
        env[i] = prev
    return env

def compress_band(band, sr, threshold_db=-24.0, ratio=2.0, attack_ms=10.0, release_ms=120.0, makeup_db=0.0):
    mono = np.mean(band, axis=1)
    env = envelope_detector(mono, sr, attack_ms, release_ms)
    thr = 10**(threshold_db/20.0)
    gain = np.ones_like(env)
    for i, e in enumerate(env):
        if e <= thr:
            g = 1.0
        else:
            over_db = 20*np.log10(e/thr)
            red_db = over_db - over_db/ratio
            g = 10**(-red_db/20.0)
        gain[i] = g
    gain *= 10**(makeup_db/20.0)
    return band * gain[:, None]

def peak_normalize(x, target_db=-1.0, eps=1e-12):
    peak = np.max(np.abs(x)) + eps
    gain = 10**(target_db/20.0) / peak
    return x * gain

def boost_bass(in_wav, out_wav,
               shelf_db=3.0, shelf_hz=80.0, shelf_slope=0.5,
               dyn_bass=False, dyn_cutoff=120.0, thr_db=-24.0, ratio=2.0, atk_ms=10.0, rel_ms=120.0, makeup_db=0.0,
               hpf_hz=20.0, headroom_db=-1.0):
    """
    Bass enhancement:
      1) Highâ€‘pass at hpf_hz to clean DC/subsonics.
      2) Lowâ€‘shelf boost (shelf_db @ shelf_hz). Start with 2â€“3 dB at 70â€“90 Hz.
      3) Optional dynamic bass: compress lows (< dyn_cutoff) to control boom.
      4) Peak-normalize to headroom_db (default -1 dBFS).

    Returns processed float32 numpy array.
    """
    sr, data = wavfile.read(in_wav)
    x = _to_float32(data)
    x = _ensure_stereo(x)

    # cleanup
    y = highpass(x, sr, hz=hpf_hz, order=2)

    # static shelf
    if shelf_db != 0.0:
        y = lowshelf_biquad(y, sr, f0=shelf_hz, gain_db=shelf_db, S=shelf_slope)

    # optional dynamic control on lows
    if dyn_bass:
        lows = band_low(y, sr, cutoff=dyn_cutoff, order=4)
        lows_c = compress_band(lows, sr, threshold_db=thr_db, ratio=ratio, attack_ms=atk_ms, release_ms=rel_ms, makeup_db=makeup_db)
        # replace only the low band delta to avoid double-boosting highs
        y = y - lows + lows_c

    # headroom
    y = np.clip(y, -1.0, 1.0)
    y = peak_normalize(y, target_db=headroom_db)

    wavfile.write(out_wav, sr, y.astype(np.float32))
    return y

# --- Quick A/B helpers ---
def compare_two(path_a, path_b, label_a="A", label_b="B"):
    def _load_mono(p):
        sr, d = wavfile.read(p)
        x = _to_float32(d)
        x = x if x.ndim == 1 else np.mean(x, axis=1)
        return sr, x

    srA, a = _load_mono(path_a)
    srB, b = _load_mono(path_b)
    # resample B to A if needed
    if srA != srB:
        gcd = np.gcd(srB, srA)
        up = srA // gcd; down = srB // gcd
        b = signal.resample_poly(b, up, down)
        srB = srA

    # Spectrum
    n = 1<<16
    Aseg = a[:min(len(a), n)]
    Bseg = b[:min(len(b), n)]
    winA = np.hanning(len(Aseg)); winB = np.hanning(len(Bseg))
    SA = np.fft.rfft(Aseg*winA); SB = np.fft.rfft(Bseg*winB)
    fA = np.fft.rfftfreq(len(Aseg), 1/srA); fB = np.fft.rfftfreq(len(Bseg), 1/srB)
    plt.figure()
    plt.plot(fA, _db(np.abs(SA)), label=label_a)
    plt.plot(fB, _db(np.abs(SB)), label=label_b)
    plt.xlim(20, 20000); plt.xscale('log')
    plt.xlabel("Frequency (Hz)"); plt.ylabel("Magnitude (dB)"); plt.title("Spectrum (log freq)")
    plt.legend()
    plt.show()

    # Low-band energy comparison
    def band_power(x, sr, f_lo, f_hi):
        sos = signal.butter(4, [f_lo/(sr*0.5), f_hi/(sr*0.5)], btype='band', output='sos')
        y = signal.sosfilt(sos, x)
        return float(np.sqrt(np.mean(y**2)))

    bassA = band_power(a, srA, 30, 120)
    bassB = band_power(b, srB, 30, 120)
    print(f"Bass RMS 30â€“120 Hz: {label_a}={bassA:.4f}, {label_b}={bassB:.4f}, Î”={bassB-bassA:+.4f}")


```

```python
# Robust version: tighten kick & bass with NaN/Inf sanitization and numerically-stable SOS filters.
# Includes guardrails for cutoff frequencies and sample-rate dependent clamping.
#
# Use:
# y = tighten_kick_bass_robust("mix.wav","mix_punchier.wav",
#                               kick_lo=40, kick_hi=110, low_cutoff=120,
#                               duck_depth_db=4.0, attack_ms=4, release_ms=90)
# sanity_check("mix.wav"); sanity_check("mix_punchier.wav")
#
# If you previously saw NaNs, this version should prevent them.

import numpy as np
from scipy.io import wavfile
from scipy import signal

def _to_float32(x):
    if x.dtype == np.int16:  return x.astype(np.float32) / 32768.0
    if x.dtype == np.int32:  return x.astype(np.float32) / 2147483648.0
    if x.dtype == np.uint8:  return (x.astype(np.float32) - 128.0) / 128.0
    return x.astype(np.float32)

def _ensure_stereo(x):
    if x.ndim == 1: return np.stack([x, x], axis=-1)
    if x.shape[1] == 1: return np.repeat(x, 2, axis=1)
    return x

def _sanitize(x):
    # Replace NaN/Inf with 0 and hard-clip extreme outliers
    x = np.nan_to_num(x, nan=0.0, posinf=0.0, neginf=0.0)
    x = np.clip(x, -4.0, 4.0)  # generous pre-clip to avoid exploding IIR states
    return x

def _safe_norm(freq, sr, lo=1.0, hi_ratio=0.95):
    # Clamp absolute Hz to a safe normalized region (0 < wn < 1)
    nyq = 0.5 * sr
    f = float(freq)
    f = max(lo, min(f, nyq * hi_ratio))
    return f

def _sos_hp(sr, hz, order=2):
    hz = _safe_norm(hz, sr)
    return signal.butter(order, hz/(sr*0.5), btype='highpass', output='sos')

def _sos_lp(sr, hz, order=4):
    hz = _safe_norm(hz, sr)
    return signal.butter(order, hz/(sr*0.5), btype='lowpass', output='sos')

def _sos_bp(sr, lo_hz, hi_hz, order=4):
    lo = _safe_norm(lo_hz, sr)
    hi = _safe_norm(hi_hz, sr)
    if hi <= lo:  # ensure valid
        hi = min(max(lo * 1.2, lo + 5.0), 0.49 * sr)
    wn = [lo/(sr*0.5), hi/(sr*0.5)]
    return signal.butter(order, wn, btype='bandpass', output='sos')

def peak_normalize(x, target_db=-1.0, eps=1e-9):
    peak = np.max(np.abs(x))
    if not np.isfinite(peak) or peak < eps:
        return np.zeros_like(x)
    return x * (10**(target_db/20.0) / peak)

def envelope_from_band_sos(band, sr, env_lp_hz=10.0):
    mono = np.mean(band, axis=1)
    mono = _sanitize(mono)
    rect = np.abs(mono)
    sos = signal.butter(2, _safe_norm(env_lp_hz, sr)/(sr*0.5), btype='lowpass', output='sos')
    env = signal.sosfilt(sos, rect)
    # Robust scaling to 0..1
    p95 = np.percentile(env[~np.isnan(env)], 95) if env.size else 0.0
    if not np.isfinite(p95) or p95 <= 1e-9:
        return np.zeros_like(env)
    env = np.clip(env / p95, 0.0, 1.0)
    env = _sanitize(env)
    return env

def smooth_sidechain(env, sr, attack_ms=4.0, release_ms=90.0):
    a_a = np.exp(-1.0/((attack_ms/1000.0)*sr))
    a_r = np.exp(-1.0/((release_ms/1000.0)*sr))
    out = np.zeros_like(env)
    prev = 0.0
    for i, e in enumerate(env):
        if e > prev:
            prev = a_a*prev + (1-a_a)*e
        else:
            prev = a_r*prev + (1-a_r)*e
        out[i] = prev
    return out

def peaking_eq_sos(x, sr, f0, gain_db, Q=1.0):
    # RBJ peaking implemented with SOS via bilinear transform
    A = 10**(gain_db/40.0)
    w0 = 2*np.pi*_safe_norm(f0, sr)/sr
    alpha = np.sin(w0)/(2*Q)
    cosw0 = np.cos(w0)
    b0 = 1 + alpha*A
    b1 = -2*cosw0
    b2 = 1 - alpha*A
    a0 = 1 + alpha/A
    a1 = -2*cosw0
    a2 = 1 - alpha/A
    b = np.array([b0, b1, b2])/a0
    a = np.array([1.0, a1/a0, a2/a0])
    # Convert to SOS for numerical stability
    sos = signal.tf2sos(b, a)
    return signal.sosfilt(sos, x, axis=0)

def tighten_kick_bass_robust(in_wav, out_wav,
                             kick_lo=40.0, kick_hi=110.0,
                             low_cutoff=120.0,
                             duck_depth_db=4.0,
                             attack_ms=4.0, release_ms=90.0,
                             mud_hz=180.0, mud_db=-1.2, mud_Q=1.0,
                             hpf_hz=20.0, headroom_db=-1.0):
    sr, data = wavfile.read(in_wav)
    x = _ensure_stereo(_to_float32(data))
    x = _sanitize(x)

    # 1) High-pass cleanup
    sos_hp = _sos_hp(sr, hpf_hz, order=2)
    y = signal.sosfilt(sos_hp, x, axis=0)

    # 2) Split bands
    sos_lp = _sos_lp(sr, low_cutoff, order=4)
    sos_kb = _sos_bp(sr, kick_lo, kick_hi, order=4)
    low = signal.sosfilt(sos_lp, y, axis=0)
    kick = signal.sosfilt(sos_kb, y, axis=0)
    nonkick_low = low - kick

    # 3) Kick envelope and duck
    env = envelope_from_band_sos(kick, sr, env_lp_hz=10.0)
    env = smooth_sidechain(env, sr, attack_ms=attack_ms, release_ms=release_ms)
    floor_gain = 10**(-duck_depth_db/20.0)
    gain_curve = floor_gain + (1.0 - floor_gain) * (1.0 - env)
    gain_curve = gain_curve[:, None]
    ducked_nonkick = nonkick_low * gain_curve

    # 4) Recombine
    lows_tight = ducked_nonkick + kick
    high = y - low
    y_out = lows_tight + high

    # 5) Optional anti-mud
    if mud_db != 0.0:
        y_out = peaking_eq_sos(y_out, sr, f0=mud_hz, gain_db=mud_db, Q=mud_Q)

    # 6) Final sanitize + headroom
    y_out = _sanitize(y_out)
    y_out = np.clip(y_out, -1.0, 1.0)
    y_out = peak_normalize(y_out, target_db=headroom_db)

    wavfile.write(out_wav, sr, y_out.astype(np.float32))
    return y_out

def sanity_check(path):
    sr, data = wavfile.read(path)
    x = _to_float32(data)
    n_nan = np.isnan(x).sum()
    n_inf = np.isinf(x).sum()
    peak = np.max(np.abs(np.nan_to_num(x, nan=0.0, posinf=0.0, neginf=0.0)))
    print(f"{path}: sr={sr}, shape={x.shape}, NaNs={n_nan}, Infs={n_inf}, peak={peak:.3f}")

# Example:
# sanity_check("mix.wav")
# tighten_kick_bass_robust("mix.wav","mix_punchier.wav")
# sanity_check("mix_punchier.wav")
```

```python
# Notebook-ready "bassier + punchier" processor (robust, SOS filters)
# - Cleans sub/DC
# - Adds bass via low-shelf
# - Tightens kick/bass via sidechain-like ducking of non-kick lows
# - Optional anti-mud peaking cut
# - Peak-normalizes for safe headroom
#
# Usage example:
# bass_and_punch("mix.wav", "mix_bass_punch.wav",
#                shelf_db=3.0, shelf_hz=80.0,
#                duck_depth_db=4.0, kick_lo=40.0, kick_hi=110.0,
#                low_cutoff=120.0, hpf_hz=20.0, mud_hz=180.0, mud_db=-1.0)


from textwrap import dedent
import numpy as np
from scipy.io import wavfile
from scipy import signal

# --------- Utilities ---------
def _to_float32(x):
    if x.dtype == np.int16:  return x.astype(np.float32) / 32768.0
    if x.dtype == np.int32:  return x.astype(np.float32) / 2147483648.0
    if x.dtype == np.uint8:  return (x.astype(np.float32) - 128.0) / 128.0
    return x.astype(np.float32)

def _ensure_stereo(x):
    if x.ndim == 1: return np.stack([x, x], axis=-1)
    if x.shape[1] == 1: return np.repeat(x, 2, axis=1)
    return x

def _sanitize(x):
    return np.clip(np.nan_to_num(x, nan=0.0, posinf=0.0, neginf=0.0), -4.0, 4.0)

def _safe_freq(freq, sr, lo=20.0, hi_ratio=0.49):
    f = float(freq)
    nyq = 0.5 * sr
    return max(lo, min(f, hi_ratio * sr))

def _sos_hp(sr, hz, order=2):
    hz = _safe_freq(hz, sr)
    return signal.butter(order, hz/(sr*0.5), btype='highpass', output='sos')

def _sos_lp(sr, hz, order=4):
    hz = _safe_freq(hz, sr)
    return signal.butter(order, hz/(sr*0.5), btype='lowpass', output='sos')

def _sos_bp(sr, lo_hz, hi_hz, order=4):
    lo = _safe_freq(lo_hz, sr)
    hi = _safe_freq(hi_hz, sr)
    if hi <= lo: hi = min(max(lo * 1.2, lo + 5.0), 0.49*sr)
    wn = [lo/(sr*0.5), hi/(sr*0.5)]
    return signal.butter(order, wn, btype='bandpass', output='sos')

def _db_to_lin(db): return 10.0**(db/20.0)

def _peak_normalize(x, target_db=-1.0, eps=1e-9):
    peak = np.max(np.abs(x))
    if not np.isfinite(peak) or peak < eps:
        return np.zeros_like(x)
    return x * (_db_to_lin(target_db) / peak)

def _envelope_from_band(band, sr, env_lp_hz=10.0):
    mono = np.mean(band, axis=1)
    mono = _sanitize(mono)
    rect = np.abs(mono)
    sos = signal.butter(2, _safe_freq(env_lp_hz, sr)/(sr*0.5), btype='lowpass', output='sos')
    env = signal.sosfilt(sos, rect)
    # Robust 0..1 scaling
    p95 = np.percentile(env, 95) if env.size else 0.0
    if not np.isfinite(p95) or p95 <= 1e-9:
        return np.zeros_like(env)
    env = np.clip(env / p95, 0.0, 1.0)
    return _sanitize(env)

def _smooth_attack_release(x, sr, attack_ms=4.0, release_ms=90.0):
    a_a = np.exp(-1.0/((attack_ms/1000.0)*sr))
    a_r = np.exp(-1.0/((release_ms/1000.0)*sr))
    y = np.zeros_like(x)
    prev = 0.0
    for i, v in enumerate(x):
        if v > prev:
            prev = a_a*prev + (1-a_a)*v
        else:
            prev = a_r*prev + (1-a_r)*v
        y[i] = prev
    return y

def _peaking_eq_sos(x, sr, f0, gain_db, Q=1.0):
    # RBJ peaking biquad -> SOS
    A = 10**(gain_db/40.0)
    w0 = 2*np.pi*_safe_freq(f0, sr)/sr
    alpha = np.sin(w0)/(2*Q)
    cosw0 = np.cos(w0)
    b0 = 1 + alpha*A
    b1 = -2*cosw0
    b2 = 1 - alpha*A
    a0 = 1 + alpha/A
    a1 = -2*cosw0
    a2 = 1 - alpha/A
    b = np.array([b0, b1, b2])/a0
    a = np.array([1.0, a1/a0, a2/a0])
    sos = signal.tf2sos(b, a)
    return signal.sosfilt(sos, x, axis=0)

# --------- Main processor ---------
def bass_and_punch(in_wav, out_wav,
                   shelf_db=3.0, shelf_hz=80.0,    # bassiness
                   duck_depth_db=4.0,               # punchiness (duck non-kick lows under kick)
                   kick_lo=40.0, kick_hi=110.0,
                   low_cutoff=120.0,
                   attack_ms=4.0, release_ms=90.0,
                   hpf_hz=20.0,
                   mud_hz=180.0, mud_db=-1.0, mud_Q=1.0,
                   headroom_db=-1.0):
    """
    Make a stereo mix both bassier and punchier:
      - HPF cleanup at hpf_hz
      - Low-shelf bass boost via gain on lows (achieved by gentle peaking proxy at shelf_hz)
      - Kick/bass tightening via envelope-duck of non-kick low band
      - Optional anti-mud dip around mud_hz
      - Peak normalize to headroom_db
    """
    sr, data = wavfile.read(in_wav)
    x = _ensure_stereo(_to_float32(data))
    x = _sanitize(x)

    # Cleanup DC / subsonics
    sos_hp = _sos_hp(sr, hpf_hz, order=2)
    y = signal.sosfilt(sos_hp, x, axis=0)

    # Split lows/highs
    sos_lp = _sos_lp(sr, low_cutoff, order=4)
    low = signal.sosfilt(sos_lp, y, axis=0)
    high = y - low

    # "Low-shelf" via broad peaking near shelf_hz (fast and stable)
    if shelf_db != 0.0:
        low = _peaking_eq_sos(low, sr, f0=shelf_hz, gain_db=shelf_db, Q=0.7)

    # Kick band & envelope
    sos_kb = _sos_bp(sr, kick_lo, kick_hi, order=4)
    kick = signal.sosfilt(sos_kb, y, axis=0)
    nonkick_low = low - kick

    env = _envelope_from_band(kick, sr, env_lp_hz=10.0)
    env = _smooth_attack_release(env, sr, attack_ms=attack_ms, release_ms=release_ms)

    # Sidechain: env 0..1 -> gain [floor .. 1.0]
    floor_gain = 10**(-abs(duck_depth_db)/20.0)
    gain_curve = floor_gain + (1.0 - floor_gain) * (1.0 - env)
    gain_curve = gain_curve[:, None]  # to stereo shape

    ducked_nonkick = nonkick_low * gain_curve
    lows_tight = ducked_nonkick + kick

    # Recombine with highs
    y_out = lows_tight + high

    # Anti-mud (optional small dip around 160â€“220 Hz)
    if mud_db != 0.0:
        y_out = _peaking_eq_sos(y_out, sr, f0=mud_hz, gain_db=mud_db, Q=mud_Q)

    # Final safety & headroom
    y_out = _sanitize(y_out)
    y_out = np.clip(y_out, -1.0, 1.0)
    y_out = _peak_normalize(y_out, target_db=headroom_db)

    wavfile.write(out_wav, sr, y_out.astype(np.float32))
    return y_out, sr

```
